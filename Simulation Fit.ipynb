```python
%pylab inline
import plotly
import h5py
```

```python
fid=open('Dropbox/phd/RawResults.txt','r')
x0=[]
for fin in fid.readlines():
    fsplit = fin.split()
    x0.append(fsplit)
x0=array(x0)
weibulRate=[]
weibulShape=[]
epoch=[]

for x1 in x0[:,4]:    
    weibulRate.append(float(x1))
for x2 in x0[:,5]:
    weibulShape.append(float(x2))
for x3 in x0[:,3]:    
    epoch.append(x3)
weibulRate=array(weibulRate)
weibulShape=array(weibulShape)
epoch=array(epoch)
fid.close()
```

```python
h5file = h5py.File('Dropbox/phd/fppFitExperiment1.h5','r')
print h5file.values()
epoch = array(h5file.get('epoch').value)
weibulRate = array(h5file.get('rateParameter').value)
weibulShape = array(h5file.get('shapeParameter').value)
h5file.close()
```

```python
rc('axes', linewidth = 2)
rc('xtick', labelsize = 18)
rc('ytick', labelsize = 18)
rc('figure', figsize = [10,10])
plot(weibulShape[epoch=='E1'], weibulRate[epoch=='E1'], '^r', markersize = 10)
plot(weibulShape[epoch=='E2'], weibulRate[epoch=='E2'], 'vy', markersize = 10)
plot(weibulShape[epoch=='E3'], weibulRate[epoch=='E3'], '.g', markersize = 10)
xlabel('shape parameter', fontsize = 18)
ylabel('rate parameter' , fontsize = 18)
legend(['epoch 1', 'epoch 2', 'epoch 3'], fontsize = 16)
ax = gca()
ax.set_axis_bgcolor("0.2")
savefig('Dropbox/phd/OldExperimentSimulationFits.pdf')
show()
```

```python
wordExperiment = {
"Epoch 1": {
    "shape parameter":weibulShape[epoch=='E1'],
    "rate parameter":weibulRate[epoch=='E1']},
"Epoch 2": {
    "shape parameter":weibulShape[epoch=='E2'],
    "rate parameter":weibulRate[epoch=='E2']},
"Epoch 3": {
    "shape parameter":weibulShape[epoch=='E3'],
    "rate parameter":weibulRate[epoch=='E3']}} 
```

```python
attr = wordExperiment['Epoch 1'].keys()
colours = {'Epoch 1': 'rgb(10,100,200)',
           'Epoch 2': 'rgb(200,100,10)',
           'Epoch 3': 'rgb(10,200,10)'}
data = []
for i in range(2):
    for j in range(2):
        for epoch in wordExperiment.keys():
            data.append({"name": epoch,
                         "x": wordExperiment[epoch][attr[i]],
                         "y": wordExperiment[epoch][attr[j]],
                         "type":"scatter",
                         "mode":"markers",
                         "marker": {'color': colours[epoch],
                                    'opacity': 0.7},
                         "xaxis": "x"+(str(i) if i!=0 else ''),
                         "yaxis": "y"+(str(j) if j!=0 else '')})
padding = 0.04
domains = [[i*padding + i*(1-3*padding)/2, i*padding + ((i+1)*(1-3*padding)/2)] for i in range(2)]
```

```python
layout = {
    "xaxis":{"domain":domains[0], "title":attr[0], 'zeroline':False,'showline':False,'ticks':'', 'titlefont':{'color': "rgb(67, 67, 67)"},'tickfont':{'color': 'rgb(102,102,102)'}},
    "yaxis":{"domain":domains[0], "title":attr[0], 'zeroline':False,'showline':False,'ticks':'', 'titlefont':{'color': "rgb(67, 67, 67)"},'tickfont':{'color': 'rgb(102,102,102)'}},
    "xaxis1":{"domain":domains[1], "title":attr[1], 'zeroline':False,'showline':False,'ticks':'', 'titlefont':{'color': "rgb(67, 67, 67)"},'tickfont':{'color': 'rgb(102,102,102)'}},
    "yaxis1":{"domain":domains[1], "title":attr[1], 'zeroline':False,'showline':False,'ticks':'', 'titlefont':{'color': "rgb(67, 67, 67)"},'tickfont':{'color': 'rgb(102,102,102)'}},
    "showlegend": False,
    "width": 300,
    "height": 300,
    "title": "Epoch data set",
    "titlefont": {'color':'rgb(67,67,67)', 'size': 20}}
```

```python
p = plotly.plotly('BuzzVII', 'dof92vob42', False)
p.iplot(data,layout=layout, width=850, height=750)
```
